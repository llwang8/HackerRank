```python
# Challenge 1
# Bank Accounts

import sys

def feeOrUpfront(n, k, x, d, p):
    fee1 = 0
    fee2 = d
    if not p:
        return 'fee'
    for i in range(n):
        fee1 += max(k, 1.0*x*p[i]/100)
    if fee1 > fee2:
        return 'upfront'
    else:
        return 'fee'
        

if __name__ == "__main__":
    q = int(raw_input().strip())
    for a0 in xrange(q):
        n, k, x, d = raw_input().strip().split(' ')
        n, k, x, d = [int(n), int(k), int(x), int(d)]
        p = map(int, raw_input().strip().split(' '))
        result = feeOrUpfront(n, k, x, d, p)
        print result
        
        
```

```python
# Challenge 2
# Buy Maximum stocks

import sys

def buyMaximumProducts(n, k, a):
    stocks = []
    total = 0
    count = 0
    if len(a) == 0:
        return 0
    for i in range(n):
        stocks += [a[i]] * (i+1)
    stocks.sort()
    for i in range(len(stocks)):
        if k - total < stocks[i]:
            return count 
        else:
            total += stocks[i]
            count += 1
    return count

if __name__ == "__main__":
    n = int(raw_input().strip())
    arr = map(int, raw_input().strip().split(' '))
    k = long(raw_input().strip())
    result = buyMaximumProducts(n, k, arr)
    print result

# 2nd attemp
def buyMaximumNProducts(n, k, a):
    total = 0
    days = [i for i in range(1, n+1)]
    daysPrices = zip(days, a)
    daysPrices = sorted(daysPrices, key=lambda x:x[1])
    for e in daysPrices:
        if k < e[1]:
            continue
        maxN = int(k/e[1])
        if maxN > e[0]:
            total += e[0]
            k = k - e[0]*e[1]
        elif maxN >= 1:
            total += maxN
            k = k - e[1] * maxN
    return total
```

```python
# Challenge 3
# Trade Profit
import sys

def buyMaximumProducts(n, k, a):
    stocks = []
    total = 0

    if len(a) == 0:
        return 0
    for i in range(n):
        stocks += [a[i]] * (i+1)
    stocks.sort()
    m = len(stocks)
    for i in range(m):
        if k - total < stocks[i]:
            return i 
        else:
            total += stocks[i]
    return m

if __name__ == "__main__":
    n = int(raw_input().strip())
    arr = map(int, raw_input().strip().split(' '))
    k = long(raw_input().strip())
    result = buyMaximumProducts(n, k, arr)
    print result
```

```python
# Challenge 5
# Currencies


import sys
from itertools import permutations

def currencies(n, x, s, f, m, A):
    mod = 1000000007
    
    currencies = [i for i in range(n)]
    currencies.remove(s)
    currencies.remove(f)

    exCurrencies = permutations(currencies, m-1)
        
    exRates = []
    for e in exCurrencies:
        exR = 1
        i = s
        for j in e:
            exR *= A[i][j]
            i = j
        exR *= A[i][f]
        exRates.append(exR)
    #print exRates
    #t = max(exRates) 
    
    return int((max(exRates) * x)%mod)
            

if __name__ == "__main__":
    n = int(raw_input().strip())
    x, s, f, m = raw_input().strip().split(' ')
    x, s, f, m = [int(x), int(s), int(f), int(m)]
    A = []
    for A_i in xrange(n):
        A_temp = map(float,raw_input().strip().split(' '))
        A.append(A_temp)
    
    print currencies(n, x, s, f, m, A)
```
