```python
# Challenge 3: Sliding Inversion

# 1st & 4th  attempt
import sys

def inversionCount(n, m, a):
    result = 0
    def count_inversion(subarr):
        count = 0
        for k in range(m):
            for l in range(k+1, m):
                if subarr[k] > subarr[l]:
                    count += 1
        return count
    
    for i in range(n-m+1):
        sub = a[i:m+i]
        if sub != sorted(sub):
            result += count_inversion(sub)
    
    return result         

if __name__ == "__main__":
    n, m = raw_input().strip().split(' ')
    n, m = [int(n), int(m)]
    a = map(int, raw_input().strip().split(' '))
    result = inversionCount(n, m, a)
    print result
    
    
# 2nd attempt
import sys, numpy 

def inversionCount(n, m, a):
    result = 0
    inversion_tbl = numpy.zeros((n,n), dtype=int)
    for i in range(n):
        for j in range(i+1, n):
            if a[i] > a[j]:
                inversion_tbl[i,j] = 1
    print inversion_tbl
    
    for i in range(n-m+1):
        #sub = a[i:m+i]
        result += numpy.sum(inversion_tbl[i:m+i, i:m+i])
    
    return result
        
if __name__ == "__main__":
    n, m = raw_input().strip().split(' ')
    n, m = [int(n), int(m)]
    a = map(int, raw_input().strip().split(' '))
    result = inversionCount(n, m, a)
    print result
    
# 3rd attempt
import sys

def inversionCount(n, m, a):
    result = 0
    inversion_tbl = []
    for i in range(n):
        temp = []
        for j in range(n):
            if i >= j:
                temp.append(0)
            else:
                if a[i] <= a[j]:
                    temp.append(0)
                else: 
                    temp.append(1)
        inversion_tbl.append(temp)
    #print inversion_tbl
    
    for i in range(n-m+1):
        #sub = a[i:m+i]
        result += sum([inversion_tbl[k][l] for k in range(i, m+i) for l in range(i, m+i)])
    
    return result        

if __name__ == "__main__":
    n, m = raw_input().strip().split(' ')
    n, m = [int(n), int(m)]
    a = map(int, raw_input().strip().split(' '))
    result = inversionCount(n, m, a)
    print result

# 4th attempt
import sys
from itertools import product

def inversionCount(n, m, a):
    result = 0  
    for i in range(n-m+1):   
        sub = a[i:i+m]
        #print sub
        if sub != sorted(sub):
            temp = [i for i in range(0, m)]
            indexes = list(product(temp, repeat=2))
            for item in indexes:
                k = item[0]
                l = item[1]
                if (k < l) and (sub[k] > sub[l]):
                    result += 1   
    return result         

if __name__ == "__main__":
    n, m = raw_input().strip().split(' ')
    n, m = [int(n), int(m)]
    a = map(int, raw_input().strip().split(' '))
    result = inversionCount(n, m, a)
    print result
```
